# Diffraction shift correction with py4DSTEM

The code in this notebook determines the shifts of the diffraction pattern in a 4D-STEM dataset associated with the scanning of the electron beam, then corrects those shifts, outputting a new 4D data hypercube in which the shifts have been removed.

The general procedure is as follows:
1. Get vacuum probe
2. Get shifts
    * Get raw shifts
    * Fit shifts
3. Correct the shifts

```python
import py4DSTEM

import numpy as np
import matplotlib.pyplot as plt

from scipy.ndimage.morphology import binary_dilation as binary_dilation
from scipy.ndimage.filters import gaussian_filter
from scipy.optimize import curve_fit

from time import time
```

## 1. Get the vacuum probe

The aim here is to create a convolution kernel in the shape of the unscattered electron beam, which can then be used to identify the Bragg disks in each diffraction pattern of the datacube.  We procede as follows:

* Get average vacuum probe
    * Select an ROI over (nearly) vacuum
    * Find shifts of DPs within ROI relative to one reference DP
    * Register and average all DPs in ROI
* Create vacuum probe kernel
    * Mask probe outside of central disk
    * Get probe CoM and size
    * Normalize probe to integrate to 1
    * Subtract normalized 2D gaussian to get kernel integrating to 0
    * Center kernel in Fourier space (i.e. at corners of array)

```python
# ROI

xmin,xmax=44,60
ymin,ymax=2,17

im_ROI_sel = np.average(dc.data4D,axis=(2,3))
im_ROI_sel[xmin:xmax,ymin:ymax] = 0
fig,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(12,8))
ax1.matshow(np.average(dc.data4D,axis=(2,3)),cmap='gray')
ax2.matshow(im_ROI_sel,cmap='gray')
ax3.matshow(np.average(dc.data4D,axis=(2,3))[xmin:xmax,ymin:ymax,],cmap='gray')
plt.show()
```

```python
# Average DP, unshifted

DP_ave_unshifted = np.average(dc.data4D[xmin:xmax,ymin:ymax],axis=(0,1))

fig,ax=plt.subplots()
ax.matshow(DP_ave_unshifted,cmap='gray')
plt.show()
```

```python
# Get average shifted diffraction pattern

ref_DP = dc.data4D[xmin,ymin,:,:]

for n in range(1,(xmax-xmin)*(ymax-ymin)):
    i = xmin + n%(xmax-xmin)
    j = ymin + int(n/(xmax-xmin))
    curr_DP = dc.data4D[i,j,:,:]
    
    xshift,yshift = py4DSTEM.process.utils.get_shift(ref_DP, curr_DP)
    curr_DP_shifted = py4DSTEM.process.utils.get_shifted_ar(curr_DP,xshift,yshift)
    ref_DP = ref_DP*(n-1)/n + curr_DP_shifted/n
```

```python
# Get (initial) center of mass
# Show shifted, averaged DP

xCoM_init,yCoM_init = py4DSTEM.process.utils.get_CoM(ref_DP)

fig,(ax1,ax2)=plt.subplots(1,2)
ax1.matshow(ref_DP,cmap='gray')
ax2.matshow(ref_DP[int(xCoM_init)-50:int(xCoM_init)+50,int(yCoM_init)-50:int(yCoM_init)+50])
plt.show()
```

```python
# Compare shifted to unshifted ave DPs

xCoM_unshifted,yCoM_unshifted = py4DSTEM.process.utils.get_CoM(DP_ave_unshifted)

fig,(ax1,ax2)=plt.subplots(1,2)
ax1.matshow(ref_DP[int(xCoM_init)-50:int(xCoM_init)+50,int(yCoM_init)-50:int(yCoM_init)+50])
ax2.matshow(DP_ave_unshifted[int(xCoM_unshifted)-50:int(xCoM_unshifted)+50,int(yCoM_unshifted)-50:int(yCoM_unshifted)+50])
plt.show()
```

```python
# Mask probe outside of central disk

mask_thresh = 0.2
mask_expansion = 12

mask = ref_DP > np.max(ref_DP)*mask_thresh
mask = binary_dilation(mask,iterations=mask_expansion)

#maskDist = 0.5 + 0.5*erf(maskShift + (bwdist(~maskRound) - bwdist(bwmorph(maskRound,'dilate',1))) / (0.5*maskWidth));

probe_template = ref_DP*mask
```

```python
# Get CoM
# Compare to CoM without mask

xCoM,yCoM = py4DSTEM.process.utils.get_CoM(probe_template)

fig,(ax1,ax2)=plt.subplots(1,2)
ax1.matshow(probe_template[int(xCoM)-50:int(xCoM)+50,int(yCoM)-50:int(yCoM)+50])
ax2.matshow(ref_DP[int(xCoM_init)-50:int(xCoM_init)+50,int(yCoM_init)-50:int(yCoM_init)+50])
plt.show()

print("Mask alters CoM by {} pixels.".format(np.sqrt((yCoM-yCoM_init)**2+(xCoM-xCoM_init)**2)))
```

```python
# Get probe size

qy,qx = np.meshgrid(np.arange(dc.Q_Ny),np.arange(dc.Q_Nx))
q2 = (qx-xCoM)**2 + (qy-yCoM)**2
qstd2 = np.sum(q2*probe_template*mask) / np.sum(probe_template)
```

```python
# Normalize probe to one
# Subtract off gaussian, to yield a kernel which integrates to zero
sigma_probe_scale = 5

probe_template_norm = probe_template/np.sum(probe_template)
subtr_gaussian = np.exp(-q2 / (2*qstd2*sigma_probe_scale**2))
subtr_gaussian = subtr_gaussian/np.sum(subtr_gaussian)
probe_kernel = probe_template_norm - subtr_gaussian

print("Probe kernel sums to {}.".format(np.sum(probe_kernel)))
```

```python
# Plot line profile through probe_kernel

rad=250  # Line profile length = 2*rad
width=2  # Line profile width = 2*width

line_profile = np.sum(probe_kernel[int(xCoM)-rad:int(xCoM)+rad,int(yCoM)-width:int(yCoM)+width],axis=(1))

fig,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(16,4))
ax1.matshow(probe_kernel[int(xCoM)-50:int(xCoM)+50,int(yCoM)-50:int(yCoM)+50])
ax2.plot(np.arange(len(line_profile)),line_profile)
ax3.plot(np.arange(len(line_profile)),line_profile)
#ax3.set_ylim(-0.0005,0.0005)
plt.show()
```

```python
# Shift probe kernel to center, then to corners
probe_kernel = py4DSTEM.process.utils.get_shifted_ar(probe_kernel,-xCoM,-yCoM)
```

```python
fig,ax=plt.subplots()
ax.matshow(probe_kernel,cmap='gray')
plt.show()
```

## 2. Get the diffraction shifts

In this step the relative shift of each diffraction pattern is determined, as follows:

* For one diffraction pattern, get the Bragg disk positions of unscattered beam.
    * Convolve with the probe kernel
    * Get the positions and strengths of all bragg disks by identifying local maxima
* For all DPs, get the positions of unscattered beam.
    * Loop over all DPs, get all Bragg disk positions
    * Get the average deconvolution
    * Get average central beam position, from brightest peak of average deconvolution
    * Get the central beam for each DP, as closest peak to average central beam position
* Fit diffraction shifts
    * Fit plane or 2d parabaloid to unscattered beam positions

### 2b. Get diffraction shifts for all DPs

* For all DPs, get the positions of unscattered beam.
    * Loop over all DPs, get all Bragg disk positions
    * Get the average deconvolution
    * Get average central beam position, from brightest peak of average deconvolution
    * Get the central beam for each DP, using closest peak to average central beam position

```python
# Loop over all DP's, getting all Bragg disk positions

corrPower = 0.9
sigma_gaussianFilter = 2
edge_boundary = 20
maxNumPeaks = 40
minPeakSpacing = 32

point_list_type=np.dtype([('rx',float),('ry',float),('qx',float),('qy',float),('intensity',float)])
Bragg_peak_point_list = np.array([],dtype=point_list_type)

t0=time()
probe_kernel_FT = np.conj(np.fft.fft2(probe_kernel))
for Ri in range(dc.R_Ny):
    for Rj in range(dc.R_Nx):
        print("Analyzing scan position {},{}".format(Ri,Rj))

        # Get diffraction pattern
        DP = dc.data4D[Ri,Rj,:,:]
        
        # Cross correlation     
        cc = py4DSTEM.process.utils.get_cross_correlation_fk(DP,probe_kernel_FT,corrPower)

        # Gaussian smoothing
        cc = np.maximum(cc,0)
        cc = gaussian_filter(cc, sigma_gaussianFilter)
        
        # Get maximal points
        max_points = py4DSTEM.process.utils.get_maximal_points(cc)
        
        # Remove points at edges
        max_points[:edge_boundary,:]=False
        max_points[-edge_boundary:,:]=False
        max_points[:,:edge_boundary]=False
        max_points[:,-edge_boundary:]=False    
       
        # Make point_list with structured array datatype ('rx','ry','qx','qy','intensity')
        point_list = np.array([],dtype=point_list_type)

        max_point_indices_x, max_point_indices_y = np.nonzero(max_points)
        point_intensities = cc[max_point_indices_x,max_point_indices_y]

        length = len(point_intensities)
        for i in range(length):
            new_point = (Ri,Rj,max_point_indices_x[i],max_point_indices_y[i],point_intensities[i])
            point_list = np.append(point_list, np.array(new_point,dtype=point_list_type))
        
        # Order point list, keep only up to maxNumPeaks
        numPeaks = min(maxNumPeaks,length)
        point_list = np.sort(point_list,order='intensity')[::-1]
        point_list = point_list[:numPeaks]        
        
        # Remove peaks that are too close together
        r2 = minPeakSpacing**2
        deletePeaks = np.zeros(numPeaks,dtype=bool)
        for i in range(numPeaks):
            if deletePeaks[i] == False:
                tooClose = ( (point_list['qx']-point_list['qx'][i])**2 + \
                          (point_list['qy']-point_list['qy'][i])**2 ) < r2
                tooClose[:i+1] = False
                deletePeaks[tooClose] = True
        point_list = np.delete(point_list,deletePeaks.nonzero()[0])
    
        # Add current point list to running list
        Bragg_peak_point_list = np.append(Bragg_peak_point_list, point_list)
            
t=time()-t0
print("Finished in {}h {}m {}s".format(int(t/3600),int(t/60),int(t%60)))
```

```python
# Get the average deconvolution

average_deconvolution = np.zeros((dc.Q_Ny,dc.Q_Nx))

for i in range(len(Bragg_peak_point_list)):
    qx = int(Bragg_peak_point_list['qx'][i])
    qy = int(Bragg_peak_point_list['qy'][i])
    average_deconvolution[qx,qy] += Bragg_peak_point_list['intensity'][i]
```

```python
# Smooth average deconvolution

sigma_gaussianFilter2 = 2
average_deconvolution_gf = gaussian_filter(average_deconvolution, sigma_gaussianFilter2)

fig,((ax11,ax12),(ax21,ax22))=plt.subplots(2,2,figsize=(16,16))
ax11.matshow(average_deconvolution_gf)
ax12.matshow(average_deconvolution_gf[x0-rad:x0+rad,y0-rad:y0+rad])
ax21.matshow(average_deconvolution_gf**0.25)
ax22.matshow(average_deconvolution_gf[x0-rad:x0+rad,y0-rad:y0+rad]**0.25)
plt.show()
```

```python
# Get average central beam position, from average deconvolution CoM or brightest peak

central_beam_qx0, central_beam_qy0 = py4DSTEM.process.utils.get_CoM(average_deconvolution)
    
#central_beam_qx0, central_beam_qy0 = np.unravel_index(np.argmax(average_deconvolution_gf),average_deconvolution.shape)
```

```python
# Get the diffraction shifts for each DP, using closest peak to CoM of average deconvolution

diffraction_shifts1 = np.empty((2,dc.R_Nx,dc.R_Ny))
diffraction_shifts2 = np.empty((2,dc.R_Nx,dc.R_Ny))

for Ri in range(dc.R_Nx):
    for Rj in range(dc.R_Ny):
        mask = (Bragg_peak_point_list['rx']==Rj) & (Bragg_peak_point_list['ry']==Ri)
        point_list = Bragg_peak_point_list[mask]
        
        r2 = (point_list['qx']-central_beam_qx0)**2 + (point_list['qy']-central_beam_qy0)**2
        index = np.argmin(r2)
        diffraction_shifts1[0,Ri,Rj] = point_list['qx'][index]
        diffraction_shifts1[1,Ri,Rj] = point_list['qy'][index]
        
        index = np.argmax(point_list['intensity'])
        diffraction_shifts2[0,Ri,Rj] = point_list['qx'][index]
        diffraction_shifts2[1,Ri,Rj] = point_list['qy'][index]
```

```python
# Subtract off average from diffraction shifts

#diffraction_shifts1[0,:,:]=diffraction_shifts1[0,:,:]-np.average(diffraction_shifts1[0,:,:])
#diffraction_shifts1[1,:,:]=diffraction_shifts1[1,:,:]-np.average(diffraction_shifts1[1,:,:])

#diffraction_shifts2[0,:,:]=diffraction_shifts2[0,:,:]-np.average(diffraction_shifts2[0,:,:])
#diffraction_shifts2[1,:,:]=diffraction_shifts2[1,:,:]-np.average(diffraction_shifts2[1,:,:])
```

```python
# Plot shifts

fig,((ax11,ax12),(ax21,ax22)) = plt.subplots(2,2,figsize=(18,18))
ax11.matshow(diffraction_shifts1[0,:,:],cmap="RdBu")
ax12.matshow(diffraction_shifts1[1,:,:],cmap="RdBu")
ax21.matshow(diffraction_shifts2[0,:,:],cmap="RdBu")
ax22.matshow(diffraction_shifts2[1,:,:],cmap="RdBu")
plt.show()
```

```python
# Take distance between diffraction shifts detemined from positions and intensities

diff2 = (diffraction_shifts1[0,:,:]-diffraction_shifts2[0,:,:])**2 + \
        (diffraction_shifts1[1,:,:]-diffraction_shifts2[1,:,:])**2

fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(17,5))
ax1.matshow(diff2)
ax2.hist(diff2.ravel(),bins=np.linspace(0,100000,100))
ax3.hist(diff2.ravel(),bins=np.linspace(0,100,100))
plt.show()
```

```python
# Get the diffraction shifts for each DP
# Use only peaks where the peak (1) with highest intensity and (2) closest to CoM  are the same
# Mask off others

diffraction_shifts = np.zeros((2,dc.R_Nx,dc.R_Ny))
diffraction_shifts_known = np.zeros((dc.R_Nx,dc.R_Ny),dtype=bool)

for Ri in range(dc.R_Nx):
    for Rj in range(dc.R_Ny):
        mask = (Bragg_peak_point_list['rx']==Rj) & (Bragg_peak_point_list['ry']==Ri)
        point_list = Bragg_peak_point_list[mask]
        
        r2 = (point_list['qx']-central_beam_qx0)**2 + (point_list['qy']-central_beam_qy0)**2
        index1 = np.argmin(r2)
        index2 = np.argmax(point_list['intensity'])
        
        if index1 == index2:
            diffraction_shifts[0,Ri,Rj] = point_list['qx'][index1]
            diffraction_shifts[1,Ri,Rj] = point_list['qy'][index1]
            diffraction_shifts_known[Ri,Rj] = True
```

```python
# Subtract off average from diffraction shifts

diffraction_shifts[0,:,:]=diffraction_shifts[0,:,:]-np.average(diffraction_shifts[0,diffraction_shifts_known])
diffraction_shifts[1,:,:]=diffraction_shifts[1,:,:]-np.average(diffraction_shifts[1,diffraction_shifts_known])

diffraction_shifts[0,diffraction_shifts_known==False] = 0
diffraction_shifts[1,diffraction_shifts_known==False] = 0
```

```python
# Plot known shifts

fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,18))
ax1.matshow(diffraction_shifts[0,:,:],cmap="RdBu")
ax2.matshow(diffraction_shifts[1,:,:],cmap="RdBu")
plt.show()
```

```python
# Fit with only datapoints where diffraction_shifts_known==True

fit_function = py4DSTEM.process.fit.plane

popt_x, pcov_x, diffraction_shifts_fit_x = py4DSTEM.process.fit.fit_2D(fit_function, diffraction_shifts[0,:,:], data_mask=diffraction_shifts_known)
popt_y, pcov_y, diffraction_shifts_fit_y = py4DSTEM.process.fit.fit_2D(fit_function, diffraction_shifts[1,:,:], data_mask=diffraction_shifts_known)

fig,((ax11,ax12,ax13),(ax21,ax22,ax23))=plt.subplots(2,3,figsize=(24,16))
ax11.matshow(diffraction_shifts[0,:,:],cmap="RdBu")
ax12.matshow(diffraction_shifts_fit_x,cmap="RdBu")
ax13.matshow(diffraction_shifts[0,:,:]-diffraction_shifts_fit_x,cmap="RdBu")
ax21.matshow(diffraction_shifts[1,:,:],cmap="RdBu")
ax22.matshow(diffraction_shifts_fit_y,cmap="RdBu")
ax23.matshow(diffraction_shifts[1,:,:]-diffraction_shifts_fit_y,cmap="RdBu")
plt.show()
```

## 3. Correct the diffraction shifts

Finally, the diffraction pattern at each scan position is corrected for the calculated shifts.

* Datacube 
    * Apply shifts at each beam positions
    * Save new, processed datacube
* Averaged deconvolution
    * Apply shifts to recalculated averaged deconvolution

### 3a. Correct diffraction shifts in raw datacube

* Apply shifts at each beam positions
* Save new, processed datacube

```python
# Apply shifts

data4D_corrected = np.empty_like(dc.data4D,dtype='float32')

t0=time()
for i in range(dc.R_Ny):
    for j in range(dc.R_Nx):
        print("Shifting DP from scan position {},{}".format(i,j))
        data4D_corrected[i,j,:,:] = py4DSTEM.process.utils.get_shifted_ar(dc.data4D[i,j,:,:], 
                                                   -diffraction_shifts_fit_x[j,i],
                                                   -diffraction_shifts_fit_y[j,i])

t=time()-t0
print("Finished in {}h {}m {}s".format(int(t/3600),int(t/60),int(t%60)))
```
