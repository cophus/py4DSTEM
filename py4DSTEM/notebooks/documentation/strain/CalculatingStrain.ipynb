# Interlude: Calculating, and interpreting, strain

The aim of this notebook is to discuss how to physically interpret the strain matrix. We'll approach the question by briefly considering how, exactly, strain is calculated.

The problem is: starting with a pair of reference ('unstrained') basis vectors $\mathbf{u}$, $\mathbf{v}$, and a pair of distorted basis vectors $\mathbf{u}'$, $\mathbf{v}'$, the ultimate aim is to determine the strain tensor:

\begin{equation}
\mathbf{\epsilon}
=
\begin{pmatrix}
\epsilon_{11}  &  \epsilon_{12}   \\
\epsilon_{21}  &  \epsilon_{22}
\end{pmatrix}
\end{equation}

We assume that the changes to the lattice vectors are small, so $\mathbf{\epsilon}$ is, more precisely, infinitesimal strain tensor.

### Left or right, right or left?

Let's call our initial and final basis vectors, respectively:

\begin{equation}
\mathbf{B}
=
\begin{pmatrix}
\mathbf{u}   &   \mathbf{v}
\end{pmatrix}
=
\begin{pmatrix}
u_{x}   &   v_{x}   \\
u_{y}   &   v_{y}
\end{pmatrix}
\end{equation}

\begin{equation}
\mathbf{B'}
=
\begin{pmatrix}
\mathbf{u'}   &   \mathbf{v'}
\end{pmatrix}
=
\begin{pmatrix}
u'_{x}   &   v'_{x}   \\
u'_{y}   &   v'_{y}
\end{pmatrix}
\end{equation}

We'd like to relate the two bases by some linear tranformation 

\begin{equation}
\mathbf{T}
=
\begin{pmatrix}
a   &   b   \\
c   &   d
\end{pmatrix}
\end{equation}

<img src="interpreting_strain_1.png" alt="linear combination" style="width: 400px;"/>

We have two options for how perform this calculation, namely: from the left, or from the right.  To wit:

\begin{equation}
\mathbf{B}' = \mathbf{T}^{L}\mathbf{B}
\end{equation}

or

\begin{equation}
\mathbf{B}' = \mathbf{B}\mathbf{T}^{R}
\end{equation}

Both of these are *correct*, in the sense that both will give a matrix which represents distorting basis $\mathbf{B}$ into basis $\mathbf{B}'$.  However, *how* that distortion is represented by $\mathbf{T}^L$ or $\mathbf{T}^R$ is not the same, and the two operations have rather different geometric interpretations.

### Multiplication from the right

We'll start with the transformation from the right, $\mathbf{T}_R$.  Expanding the matrix equation:

\begin{align}
\begin{pmatrix}
u_x'   &   v_x'   \\
u_y'   &   v_y'
\end{pmatrix}
&=
\begin{pmatrix}
u_x   &   v_x   \\
u_y   &   v_y
\end{pmatrix}
\begin{pmatrix}
a  &   b   \\
c  &   d
\end{pmatrix}  \\
&=
\begin{pmatrix}
a u_x + c v_x   &   b u_x + d v_x   \\
a u_y + c v_y   &   b u_y + d v_y
\end{pmatrix}
\end{align}

The final vectors then become:

\begin{align}
\begin{pmatrix}
u_x' \\
u_y'
\end{pmatrix}
=
a
\begin{pmatrix}
u_x \\
u_y
\end{pmatrix}
+
c
\begin{pmatrix}
v_x \\
v_y
\end{pmatrix}
\end{align}

\begin{equation}
\begin{pmatrix}
v_x' \\
v_y'
\end{pmatrix}
=
b
\begin{pmatrix}
u_x \\
u_y
\end{pmatrix}
+
d
\begin{pmatrix}
v_x \\
v_y
\end{pmatrix}
\end{equation}

Or, more succinctly:

\begin{equation}
\mathbf{u}' = a\mathbf{u} + c\mathbf{v} \\
\mathbf{v}' = b\mathbf{u} + d\mathbf{v}
\end{equation}

#### Interpreting $\mathbf{T}_R$: Decomposition of $\mathbf{B}'$ in terms of $\mathbf{B}$

The last equation above is depicted geometrically below.  The interpretation is: $\mathbf{T}_R$ can be thought of as a decomposition of the final lattice vectors in terms of initial lattice vectors.  The equation $\mathbf{u}' = a\mathbf{u} + c\mathbf{v}$ says that we can think of $\mathbf{u}'$ as the sum of a vector along $\mathbf{u}$ (in red) and a vector along $\mathbf{v}$ (blue).  Similarly, $\mathbf{v}'$ is represented here as the sum of vectors along $\mathbf{v}$ (green) and $\mathbf{u}$ (purple). 

<img src="interpreting_strain_2.png" alt="linear combination" style="width: 600px;"/>

If the basis vectors are orthogonal or close to orthogonal, then the stretch of $\mathbf{u}$ lives in $a$, and the stretch of $\mathbf{v}$ lives in $d$.  Letting $a=1+\epsilon_{uu}$ and $d=1+\epsilon_{vv}$, the quantities $\epsilon_{uu}$ and $\epsilon_{vv}$ are directly interpretable as the compression/tension of the basis vectors, as percentages of the original vector lengths.  As long as the (pseudo)-orthogonality condition holds, this is a clean, simple interpretation of compression/tension.

If the basis vectors are not orthogonal, then the off diagonal terms will begin to contribute to the stretching or compression, complicating this simple interpretation.  But there's an even worse problem here, and it has to do with how we interpret the shearing.

#### Ambiguity in the shear strain

Even assuming near-orthogonal lattice vectors, interpretation of the off diagonal components is somewhat messy.  The off-diagonal elements $b$ and $c$ represent the shearing (and rotation, but let's set that aside for the moment) of the unit cell.  For compression/tension, we landed on an unambiguous physical interpretation (under certain conditions!) - the percent change in the vector lengths.  Attempting something similar for the shear strain quickly runs into problems.

Consider the simple example of orthogonal basis vectors, with some small shearing $x$ of one of the two vectors.  We'll leave it unspecified for the moment which vector is being sheared; so the transformation $\mathbf{T}_R$ is either

\begin{equation}
\mathbf{T}_1
=
\begin{pmatrix}
1   &   x   \\
0   &   1
\end{pmatrix}
\,\,\,\,\,\,\,\,\,\,
\text{or}
\,\,\,\,\,\,\,\,\,\,
\mathbf{T}_2
=
\begin{pmatrix}
1   &   0   \\
x   &   1
\end{pmatrix}
\end{equation}

These transformations are shown in the figure below.  

<img src="interpreting_strain_3.png" alt="linear combination" style="width: 400px;"/>

Here, we see that $\mathbf{T}_1$ represents a shearing of $\mathbf{v}'$, and $\mathbf{T}_2$ represents a shearing of $\mathbf{u}'$, because they respectively correspond to

\begin{align}
\begin{matrix}
\mathbf{u}' = &\mathbf{u}  \\
\mathbf{v}' = &x\mathbf{u} + \mathbf{v}
\end{matrix}
\,\,\,\,\,\,\,\,\,\,
\text{and}
\,\,\,\,\,\,\,\,\,\,
\begin{matrix}
\mathbf{u}' = &\mathbf{u} + x\mathbf{v} \\
\mathbf{v}' = &\mathbf{v}
\end{matrix}
\end{align}

We're interested in how much each of these transformations shears the unit cell.  We'll quantify this by the change in the angle between the basis vectors, $\Delta\theta$.  As we're dealing with infinitesimal strain, we can take $\Delta\theta$ to be small, therefore:

\begin{align}
\begin{matrix}
\Delta\theta_1 &\approx \tan{\Delta\theta_1}   \\
               &= x \frac{\lvert\mathbf{u}\rvert}{\lvert\mathbf{v}\rvert}
\end{matrix}
\,\,\,\,\,\,\,\,\,\,\,\,
\text{and}
\,\,\,\,\,\,\,\,\,\,\,\,
\begin{matrix}
\Delta\theta_2 &\approx \tan{\Delta\theta_2}   \\
               &= x \frac{\lvert\mathbf{v}\rvert}{\lvert\mathbf{u}\rvert}
\end{matrix}
\end{align}

If we have an orthonormal basis, $\lvert\mathbf{u}\rvert = \lvert\mathbf{v}\rvert$, so $\Delta\theta_1 = \Delta\theta_2$, and all is well.  Otherwise, we have a problem: for some a transformation $\mathbf{T}$ with some shear component $x$, how much has the unit cell been distorted?  The answer is: anywhere in the range $x \frac{\lvert\mathbf{u}\rvert}{\lvert\mathbf{v}\rvert} \leq \Delta\theta \leq x \frac{\lvert\mathbf{v}\rvert}{\lvert\mathbf{u}\rvert}$.  This can be a significant ambiguity: if $x=0.01$ in, say, graphite, where $\frac{c}{a} \approx 2.7$, then the angular deformation of the unit cell is anywhere in the range of 3.7 to 27 milliradians, spanning nearly an order of magnitude!

Life would be much simpler with an orthonormal basis.  Fortunately, all our dreams can come true, if we only perform the transformation from the right side.  Wait, sorry, I mean the left side...

### Multiplication from the left

We proceed as before for multiplication from the left, $\mathbf{T}_L$, by expanding the matrix equation:

\begin{align}
\begin{pmatrix}
u_x'   &   v_x'   \\
u_y'   &   v_y'
\end{pmatrix}
&=
\begin{pmatrix}
a  &   b   \\
c  &   d
\end{pmatrix}
\begin{pmatrix}
u_x   &   v_x   \\
u_y   &   v_y
\end{pmatrix}  \\
&=
\begin{pmatrix}
a u_x + b u_y   &   a v_x + b v_y   \\
c u_x + d u_y   &   c v_x + d v_y
\end{pmatrix}
\end{align}

The final vectors then become:

\begin{equation}
\begin{pmatrix}
u_x' \\
u_y'
\end{pmatrix}
=
\begin{pmatrix}
a u_x + b u_y\\
c u_x + d u_y
\end{pmatrix}
\end{equation}

\begin{equation}
\begin{pmatrix}
v_x' \\
v_y'
\end{pmatrix}
=
\begin{pmatrix}
a v_x + b v_y\\
c v_x + d v_y
\end{pmatrix}
\end{equation}

We immediately notice something nice about the equations above: each transformed lattice vector is now expressed solely in terms of the corresponding initial vector.  That is, now $\mathbf{u}'$ is given entirely in terms of $\mathbf{u}$, and $\mathbf{v}'$ is given entirely in terms of $\mathbf{v}$.  This means that, while the physical meaning of $\mathbf{T}_R$ was tied to what the lattice vectors themselves are, geometric interpretation of $\mathbf{T}_L$ is entirely independent of the basis vectors themselves.  $\mathbf{T}_R$ expressed a decomposition of the two basis vectors in terms of each other; $\mathbf{T}_L$ expresses a transformation of a vector space. 

#### Interpreting $\mathbf{T}_L$: Transforming the space (rather than the vectors)

$\mathbf{T}_L$ can be applied to any vector in the Euclidean plane, and its geometric meaning will be the same.  Above we wrote the transformation of the two basis vectors in a single matrix multiplication, however, we could express the same tranformation by applying $\mathbf{T}_L$ to the basis vectors one by one:

\begin{equation}
\begin{pmatrix}
u_x' \\
u_y'
\end{pmatrix}
=
\begin{pmatrix}
a  &  b   \\
c  &  d
\end{pmatrix}
\begin{pmatrix}
u_x \\
u_y
\end{pmatrix}
=
\begin{pmatrix}
a u_x + b u_y\\
c u_x + d u_y
\end{pmatrix}
\end{equation}

\begin{equation}
\begin{pmatrix}
v_x' \\
v_y'
\end{pmatrix}
=
\begin{pmatrix}
a  &  b   \\
c  &  d
\end{pmatrix}
\begin{pmatrix}
v_x \\
v_y
\end{pmatrix}
=
\begin{pmatrix}
a v_x + b v_y\\
c v_x + d v_y
\end{pmatrix}
\end{equation}

In contrast, $\mathbf{T}_R$ can't be applied to a single vector at all.  The expression

\begin{equation}
\begin{pmatrix}
v_x' \\
v_y'
\end{pmatrix}
=
\begin{pmatrix}
v_x \\
v_y
\end{pmatrix}
\begin{pmatrix}
a  &  b   \\
c  &  d
\end{pmatrix}
=
???
\end{equation}

isn't meaningful!

This all means that in interpreting the geometric meaning of $\mathbf{T}_L$ we can, without any loss of generality, consider its effect on a single lattice vector.  In turn, in considering the operation of $\mathbf{T}_L$ on a generic vector $\mathbf{u}$, we can make use of the linearity; that is, for some $\mathbf{u} = c_1\mathbf{u}_1 + c_2\mathbf{u}_2$:

\begin{align}
\mathbf{T}\mathbf{u} &= \mathbf{T}\left(c_1\mathbf{u}_1 + c_2\mathbf{u}_2\right) \\
                     &= c_1\mathbf{T}\mathbf{u}_1 + c_2\mathbf{T}\mathbf{u}_2
\end{align}

All we really need to understand, then, is the effect of $\mathbf{T}_L$ on the orthonormal coordinate vectors, $\mathbf{\widehat{e}}_x$ and $\mathbf{\widehat{e}}_y$.  We can then construct any other vector by linearity.

Formally, we want to calculate the transformation of an orthonormal coordinate basis $\mathbf{B} = \left(\mathbf{e}_1 \,\,\, \mathbf{e}_2\right) = \left(\mathbf{\widehat{e}}_x \,\,\, \mathbf{\widehat{e}}_y\right)$ into some new coordinate basis $\mathbf{B}' = \left(\mathbf{e}'_1 \,\,\, \mathbf{e}'_2\right)$ under the transformation $\mathbf{T}$.  We find

\begin{align}
\mathbf{e}'_1 &= \mathbf{T}\mathbf{\widehat{e}}_x \\
\mathbf{e}'_2 &= \mathbf{T}\mathbf{\widehat{e}}_y
\end{align}

or

\begin{equation}
\begin{pmatrix}
e_{1,x}' \\
e_{1,y}'
\end{pmatrix}
=
\begin{pmatrix}
a  &  b \\
c  &  d 
\end{pmatrix}
\begin{pmatrix}
1\\
0
\end{pmatrix}
=
\begin{pmatrix}
a\\
c
\end{pmatrix}
\end{equation}

\begin{equation}
\begin{pmatrix}
e_{2,x}' \\
e_{2,y}'
\end{pmatrix}
=
\begin{pmatrix}
a  &  b \\
c  &  d 
\end{pmatrix}
\begin{pmatrix}
0\\
1
\end{pmatrix}
=
\begin{pmatrix}
b\\
d
\end{pmatrix}
\end{equation}

This is depicted in the figure below.  We see immediately that $a$ and $d$ represent stretching along the $x$ and $y$ directions, respectively, while $b$ and $c$ represent shearing, where $b$ is the shearing of $\mathbf{\widehat{e}}_y$ in the direction of $\mathbf{\widehat{e}}_x$, and $c$ is the shearing of $\mathbf{\widehat{e}}_x$ in the direction of $\mathbf{\widehat{e}}_y$.

<img src="interpreting_strain_4.png" alt="linear combination" style="width: 300px;"/>

This description captures the behavior of all of the possible vectors in $\mathcal{R}^2$ all at once, because it describes the deformation of the space itself.  This is shown in the figure below.  On the left hand side, we see the transformation $\mathbf{T}_L$ applied to the Cartesian coordinate grid (black), yielding a deformed grid (gray).

<img src="interpreting_strain_5.png" alt="linear combination" style="width: 600px;"/>

On the right, we see that the deformation of any given vector follows from linearity.  That is, for some vector $\mathbf{u} = c_1\mathbf{\widehat{e}}_x + c_2\mathbf{\widehat{e}}_y$, we can immediately write down $\mathbf{u}'=\mathbf{T}\mathbf{u}$:

\begin{align}
\mathbf{u}' &= \mathbf{T}\mathbf{u}  \\
            &= c_1\mathbf{T}\mathbf{\widehat{e}}_x + c_2\mathbf{T}\mathbf{\widehat{e}}_y  \\
            &= c_1\mathbf{e}'_1 + c_2\mathbf{e}'_2
\end{align}

where

\begin{equation}
\mathbf{e}'_1 = \mathbf{T}\mathbf{\widehat{e}}_x
=
\begin{pmatrix}
a\\
c
\end{pmatrix}
\end{equation}

\begin{equation}
\mathbf{e}'_2 = \mathbf{T}\mathbf{\widehat{e}}_y
=
\begin{pmatrix}
b\\
d
\end{pmatrix}
\end{equation}

This is illustrated in the figure for $(c_1, c_2)$ values of $(1,4)$, $(4,3)$, and $(3,1)$ in red, green, and blue, respectively.

## Calculating strain: from $\mathbf{T}$ to $\mathbf{\epsilon}$

This is all well and good, but what we really want isn't the transformation $\mathbf{T}$ - its the strain matrix, $\mathbf{\epsilon}$.  How do we relate these?

Let's start by formally writing down $\mathbf{\epsilon}$.  Let $\mathbf{B} = \left(\mathbf{e}_1 \,\,\, \mathbf{e}_2\right)$ be a set of coordinate basis vectors, and $\mathbf{B}' = \left(\mathbf{e}'_1 \,\,\, \mathbf{e}'_2\right)$ be their final, deformed counterparts.  To simplify notation, we'll further let axes of our coordinate system $(x,y)\to(x_1,x_2)$.  Then the infinitesimal strain matrix is given by:

\begin{equation}
\mathbf{\epsilon}
=
\begin{pmatrix}
\epsilon_{11}   &   \epsilon_{12}   \\
\epsilon_{21}   &   \epsilon_{22}
\end{pmatrix}
=
\begin{pmatrix}
\frac{\partial e_1}{\partial x_1}   &   \frac{\partial e_1}{\partial x_2}   \\
\frac{\partial e_2}{\partial x_1}   &   \frac{\partial e_2}{\partial x_2}
\end{pmatrix}
\end{equation}

or, more succinctly, $\epsilon_{ij} = \frac{\partial e_i}{\partial x_j}$.

To calculate the derivatives, consider the deformation from $\mathbf{B}$ to $\mathbf{B}'$ of an initially square area of side length $l$, shown below on the left.  The $\frac{\partial e_i}{\partial x_j}$ are the changes in the $\mathbf{B}$ vectors in each direction as $l \to 0$.  That is

\begin{align}
\frac{\partial e_i}{\partial x_j} &\equiv \lim_{l\to 0} \frac{(le'_i-le_i)\cdot\mathbf{\widehat{e}}_{x_j}}{l} \\
                                  &= (e'_i-e_i)\cdot\mathbf{\widehat{e}}_{x_j}
\end{align}

The factors of $l$ vanish, and inspecting the geometry of the deformation of a unit square, below on the right, gives the derivatives:

\begin{equation}
\mathbf{\epsilon}
=
\begin{pmatrix}
\epsilon_{11}   &   \epsilon_{12}   \\
\epsilon_{21}   &   \epsilon_{22}
\end{pmatrix}
=
\begin{pmatrix}
a-1   &   b   \\
c     &   d-1
\end{pmatrix}
\end{equation}

<img src="interpreting_strain_6.png" alt="linear combination" style="width: 700px;"/>

### Stretch and shear

The diagonal elements of $\mathbf{\epsilon}$ correspond to stretching. $\epsilon_{ii}$ is the compression or tension, for positive or negative values, respectively, along the $x_i$-axis.

The off-diagonal elements correspond to shearing. Because the strains here are infinitesimal, the small angle approximation allows us to interpret the $\epsilon_{ij}$'th element ($i \neq j$) directly as the angle by which the $x_i$ axis has been deflected towards the $x_j$ axis.  Refering to the last image above:

\begin{align}
\begin{matrix}
\Delta\theta_1 &\approx \tan{\Delta\theta_1}   \\
               &= b
\end{matrix}
\,\,\,\,\,\,\,\,\,\,\,\,
\text{and}
\,\,\,\,\,\,\,\,\,\,\,\,
\begin{matrix}
\Delta\theta_2 &\approx \tan{\Delta\theta_2}   \\
               &= c
\end{matrix}
\end{align}

### Symmetrizing $\mathbf{\epsilon}$: rotation and shear

We know now that the off-diagonal components $b$ and $c$ correspond to shearing; however, physically, this is still not terribly useful, because we might have very different amounts of shearing of each basis vector.  That means the total deformation of the unit cell is not immediately clear from the values of $b$ and $c$.

For instance, what happens if $c = -b$?  As shown in the image below, under this transformation, a square of unit area hasn't really been sheared at all - instead, it's been rotated.

<img src="interpreting_strain_7.png" alt="linear combination" style="width: 600px;"/>

What we really want is to decouple shearing and rotation.  That is, for some little square of unit area of our material, we want to know how much the square has been sheared, meaning how much the angle between deformed axes deviate from their initial value of 90 degrees.  And we also want to know how much the axes have rotated.

The total shear deformation is easy: we just add up the angular offsets of each of the axes.  This quantity is sometimes called the engineering shear strain, denoted here as $\gamma_{xy}$.  So $\gamma_{xy} = \Delta\theta_x + \Delta\theta_y = b + c$.  In the example above where $c=-b$, as we expect, $\gamma_{xy}=0$.

We can define the rotation as the change in angle of a ray bisecting the initial and final coordinate axes, given by $\theta = \frac{1}{2}(c-b)$.

```python
# Get strains

strain_maps = RealSlice(data=np.empty((4,dc.R_Nx,dc.R_Ny)),
                       parentDataCube=dc,
                       slicelabels=('e_uu','e_vv','e_uv','theta'),
                       name="strain_maps")

for Rx in range(dc.R_Nx):
    for Ry in range(dc.R_Ny):
        uv = np.array([[uv_map.data['ux'][Rx,Ry],uv_map.data['uy'][Rx,Ry]],
                       [uv_map.data['vx'][Rx,Ry],uv_map.data['vy'][Rx,Ry]]])
        m = lstsq(uv_ref,uv,rcond=None)[0]
        
        strain_maps.data['e_uu'][Rx,Ry] = 1 - m[0,0]
        strain_maps.data['e_vv'][Rx,Ry] = 1 - m[1,1]
        strain_maps.data['e_uv'][Rx,Ry] = -(m[0,1]+m[1,0])/2.
        strain_maps.data['theta'][Rx,Ry] = (m[0,1]-m[1,0])/2.
```

```python
# Plot

n_stds=2.5

e_uu_ave, e_uu_std = np.average(strain_maps.data['e_uu'][mask]),np.std(strain_maps.data['e_uu'][mask])
e_vv_ave, e_vv_std = np.average(strain_maps.data['e_vv'][mask]),np.std(strain_maps.data['e_vv'][mask])
e_uv_ave, e_uv_std = np.average(strain_maps.data['e_uv'][mask]),np.std(strain_maps.data['e_uv'][mask])
theta_ave, theta_std = np.average(strain_maps.data['theta'][mask]),np.std(strain_maps.data['theta'][mask])

fig,((ax11,ax12),(ax21,ax22)) = plt.subplots(2,2,figsize=(16,16))
ax11.matshow(strain_maps.data['e_uu'],vmin=-n_stds*e_uu_std, vmax=n_stds*e_uu_std, cmap="RdBu")
ax12.matshow(strain_maps.data['e_vv'],vmin=-n_stds*e_vv_std, vmax=n_stds*e_vv_std, cmap="RdBu")
ax21.matshow(strain_maps.data['e_uv'],vmin=-n_stds*e_uv_std, vmax=n_stds*e_uv_std, cmap="RdBu")
ax22.matshow(strain_maps.data['theta'],vmin=theta_ave-n_stds*theta_std, vmax=theta_ave+n_stds*theta_std, cmap="RdBu")
plt.show()
```

```python
# Plot, with gaussian smoothing

sigma = .8
n_stds=3

e_uu_smoothed = np.where(mask,gaussian_filter(np.where(mask, strain_maps.data['e_uu'], medfilt(np.where(mask,strain_maps.data['e_uu'],0),kernel_size=3)),sigma),0)
e_vv_smoothed = np.where(mask,gaussian_filter(np.where(mask, strain_maps.data['e_vv'], medfilt(np.where(mask,strain_maps.data['e_vv'],0),kernel_size=3)),sigma),0)
e_uv_smoothed = np.where(mask,gaussian_filter(np.where(mask, strain_maps.data['e_uv'], medfilt(np.where(mask,strain_maps.data['e_uv'],0),kernel_size=3)),sigma),0)
theta_smoothed = np.where(mask,gaussian_filter(np.where(mask, strain_maps.data['theta'], medfilt(np.where(mask,strain_maps.data['theta'],0),kernel_size=3)),sigma),0)

e_uu_ave, e_uu_std = np.average(e_uu_smoothed[mask]),np.std(e_uu_smoothed[mask])
e_vv_ave, e_vv_std = np.average(e_vv_smoothed[mask]),np.std(e_vv_smoothed[mask])
e_uv_ave, e_uv_std = np.average(e_uv_smoothed[mask]),np.std(e_uv_smoothed[mask])
theta_ave, theta_std = np.average(theta_smoothed[mask]),np.std(theta_smoothed[mask])

fig,((ax11,ax12),(ax21,ax22)) = plt.subplots(2,2,figsize=(16,16))
ax11.matshow(e_uu_smoothed,vmin=-n_stds*e_uu_std, vmax=n_stds*e_uu_std, cmap="RdBu")
ax12.matshow(e_vv_smoothed,vmin=-n_stds*e_vv_std, vmax=n_stds*e_vv_std, cmap="RdBu")
ax21.matshow(e_uv_smoothed,vmin=-n_stds*e_uv_std, vmax=n_stds*e_uv_std, cmap="RdBu")
ax22.matshow(theta_smoothed,vmin=theta_ave-n_stds*theta_std, vmax=theta_ave+n_stds*theta_std, cmap="RdBu")
plt.show()
```
