# Module: braggdiskdetection

This module contains functions finding the positions of the Bragg disks in a 4DSTEM scan.  Generally this will involve two steps: getting a vacuum probe, then finding the Bragg disks using the vacuum probe as a template. 

## Submodule: braggvectormap

The notebook demos functions for calculating the Bragg vector map from a PointListArray of bragg peak positions.

```python
import numpy as np
import matplotlib.pyplot as plt
import py4DSTEM

from py4DSTEM.process.braggdiskdetection import get_bragg_vector_map, get_bragg_vector_maxima_map
```

```python
# Get data

fp = "/home/ben/Data/20180905_FePO4_unlithiated/processing/Stack1_processing_temp.h5"
browser = py4DSTEM.file.io.FileBrowser(fp)
pointlistarray = browser.get_dataobject('braggpeaks')
browser.close()

Q_Nx,Q_Ny = 512,512
```

#### Bragg vector mapp

```python
# Get Bragg vector map
braggvectormap = get_bragg_vector_map(pointlistarray,Q_Nx,Q_Ny)

# Show
fig,ax = plt.subplots(figsize=(8,8))
ax.matshow(braggvectormap**0.25,cmap='jet')
plt.show()
```

```python
# Get Bragg vector maxima map
braggvectormap = get_bragg_vector_maxima_map(pointlistarray,Q_Nx,Q_Ny)

# Show
fig,ax = plt.subplots(figsize=(8,8))
ax.matshow(braggvectormap**0.25,cmap='jet')
plt.show()
```
