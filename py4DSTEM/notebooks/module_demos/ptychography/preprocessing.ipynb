### Rebin data if necessary 

```python
def rebin(a, s, mode='sum'):
    '''rebin ndarray data into a smaller ndarray of the same rank whose dimensions
    are factors of the original dimensions. eg. An array with 6 columns and 4 rows
    can be reduced to have 6,3,2 or 1 columns and 4,2 or 1 rows.
    example usages:
    >>> a=rand(6,4); b=rebin(a,(3,2))
    >>> a=rand(6); b=rebin(a,2)
    '''
    shape = np.asarray(a.shape)
    lenShape = len(shape)
    args = np.ones_like(shape) * np.asarray(s)
    factor = shape // args
    s1 = tuple()
    for i in range(lenShape):
        s1 += (factor[i],)
        s1 += (args[i],)
    ret = a.reshape(s1)
    for i in range(lenShape):
        j = (lenShape - i) * 2 - 1
        ret = ret.sum(j)
    if mode == 'mean':
        ret /= np.prod(args)
    return ret

d1 = rebin(d, (1, 1, 2, 2))

## are you loading a matlab file? maybe you need to transpose some axes then
# d1 = np.transpose(d1,(2,3,0,1))

print(f"d1.shape: {d1.shape}")
```

### set acceleration voltage in eV, convergence angle and real-space pixel size

```python
E = 80e3
lam = electron_wavelength_angstrom(E)
alpha_rad = 24.8e-3
dx = [0.205, 0.205]
```

### determine bright-field disk radius and crop to bright-field disk

```python
from py4DSTEM.process.calibration import get_probe_size

dp_mean = np.mean(d1, (0, 1))
plot(dp_mean)
r, y0, x0 = get_probe_size(dp_mean)
print(r, x0, y0)
r = np.round(r)
x0 = np.round(x0)
y0 = np.round(y0)
print(r, x0, y0)

f = 1.05
s = int(r * f)
yc = int(y0)
xc = int(x0)
dcrop = d1[:, :, yc - s:yc + s, xc - s:xc + s]

print(f"dcrop.shape: {dcrop.shape}")
dp_mean = np.mean(dcrop, (0, 1))
plot(dp_mean, 'dcrop')
```

### determine maximum scattering vector -> minimum real-space resolution

```python
rmax = dcrop.shape[2]//2

alpha_max = rmax / r * alpha_rad
print(f"alpha_max = {alpha_max}")
r_min = lam / (2*alpha_max)
k_max = [r_min, r_min]
print(f"k_max = {k_max}*A")
```

```python
import h5py
from py4DSTEM.file.io.write import transfer_metadata_dict
def datacube_to_py4DSTEM_viewable(datacube, filename, E, k_max, dxy, alpha_rad, rot):
    from os.path import splitext
    with h5py.File(splitext(filename)[0] + '.h5', 'w') as f:
        f.attrs['version_major'] = 0
        f.attrs['version_minor'] = 3
        grp = f.create_group('/4DSTEM_experiment/data/diffractionslices')
        grp = f.create_group('/4DSTEM_experiment/data/realslices')
        grp = f.create_group('/4DSTEM_experiment/data/pointlists')
        grp = f.create_group('/4DSTEM_experiment/data/pointlistarrays')
        grp = f.create_group('/4DSTEM_experiment/data/datacubes')
        grp.attrs['emd_group_type'] = 1
        grp.create_dataset('datacube_0/datacube', shape=datacube.shape, data=datacube, dtype=datacube.dtype)
        gpr = f.create_group('4D-STEM_data/metadata/original/shortlist')
        gpr = f.create_group('4D-STEM_data/metadata/original/all')
        microscope = f.create_group('4D-STEM_data/metadata/microscope')
        microscope_meta = Param()
        microscope_meta.accelerating_voltage = E
        microscope_meta.convergence_semiangle_mrad = alpha_rad * 1e3
        transfer_metadata_dict(microscope_meta, microscope)
        gpr = f.create_group('4D-STEM_data/metadata/sample')
        gpr = f.create_group('4D-STEM_data/metadata/user')
        gpr = f.create_group('4D-STEM_data/metadata/processing')
        calibration = f.create_group('4D-STEM_data/metadata/calibration')
        calibration_meta = Param()
        calibration_meta.R_pix_size = dxy
        calibration_meta.R_pix_units = 'angstrom'
        calibration_meta.K_pix_size = k_max
        calibration_meta.K_pix_units = 'angstrom^-1'
        calibration_meta.R_to_K_rotation_degrees = rot
        transfer_metadata_dict(calibration_meta, calibration)
        gpr = f.create_group('4D-STEM_data/metadata/comments')
```
